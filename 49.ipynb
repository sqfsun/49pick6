```python
import numpy as np
import pandas as pd
from random import sample
import time
```

```python
#Constant
NumList=range(1,50,1)
number_of_items = 6
picked=[]
bingo_num=[]
pick=[]
```

```python
def pick6():
    pick=sample(NumList, number_of_items)
    pick.sort()
    if pick in picked:
        pass
    else:
        picked.append(pick)
    return pick,picked
    
```

```python
def Bingo(pick):
    tenp=0
    bingoflag=0
    for p in pick:
        tenp=(p//10)+tenp
        
    if((tenp%10)==3 and sum(pick)==155):
        bingoflag=1

    return bingoflag
```

```python
def dedup():
    seen=[]
    for x in picked:
        if x not in seen:
            seen.append(x)
            return seen
```

```python
counter=0
tic=time.clock()
while len(picked)<13983816: #Enumerate C49_6=13983816
    
    pick,picked=pick6()
    bingoflag=Bingo(pick)
    if bingoflag==1:
        bingo_num.append(pick)
    counter=counter+1
    toc=time.clock
    comp=counter*100/13983816
    print(counter+":"+str(comp)+"% completed")
    ETA=(toc-tic)/(comp/100)/60
    print(ETA+" min to complete")
    
```
